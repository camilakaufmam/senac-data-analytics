### Exercicio 1

Lista para entrar na festa. Caso o usuario tenha uma idade igual ou maior de 18 anos ele deve entrar na festa caso nao eu quero que o console me sinalize!

```python
lista_de_idade = [
    { "name": "Caio", "age": 25 },
    { "name": "Fulano", "age": 17 },
    { "name": "Ciclano", "age": 18 },
    { "name": "Clebinho", "age": 21 },
]

idade = 18

lista_validada = []

for item in lista_de_idade:
    if idade <= item["age"]:
        lista_validada.append(item)
    
print(f" Podem entrar {lista_validada}")

```

### Exercicio 2

Lista para aposentadoria. Caso o usuario tenha uma idade igual ou maior de 65 anos e tenha CPF cadastrado Ele pode se aposentar.
Caso não, ele deve esta na lista de espera

```python
lista_de_pessoas = [
    { "name": "Caio", "age": 67, "cpf": "99999999999" },
    { "name": "Fulano", "age": 51, "cpf": "99999999999" },
    { "name": "Ciclano", "age": 65, "cpf": "99999999999" },
    { "name": "Clebinho", "age": 67, "cpf": None },
]

# Resultado Final
# Caio e Ciclano se aposentam
# Fulano e Clebinho não se aposenta

#obs.: Vou usar FOR pois é um loop. Usar if pois vou comparar; criar uma variavel para esse novo Array; criar uma variavel para usar no For


idade = 65
idade_invalida = []
idade_validada = []

for item in lista_de_pessoas:
  if idade <= item["age"]:
    if type(item["cpf"]) == str:
        idade_validada.append(item)
    else:
       idade_invalida.append(item)
  else:
     idade_invalida.append(item)


print(f"Vai se aposentar: {idade_validada}")

print(f"NÃO Vai se aposentar: {idade_invalida}")

```

### Exercicio 3

Lista para viagem. Em uma firma eu preciso saber qual cliente fez o pagamento para ir na viagem e tambem preciso saber qual cliente nao fez para eu conseguir cobrar ele.
Alem disso eu quero separa os clientes que pagaram em dinheiro e os que pagaram no cartao.

```python
lista_de_pessoas = [
    { "name": "Caio", "age": 27, "cpf": "99999999999", "isPay": True, "method": "Dinheiro" },
    { "name": "Fulano", "age": 51, "cpf": "99999999999", "isPay": False, "method": None },
    { "name": "Ciclano", "age": 65, "cpf": "99999999999", "isPay": True, "method": "Cartao" },
    { "name": "Clebinho", "age": 67, "cpf": None, "isPay": False, "method": "Dinheiro" },
]

# Resultado
# Caio e Ciclano: VÃO viajar
# Fulano e Clebinho NÃO pagaram
# Caio pagou no dinheiro
# Cliclano pagou no Cartão

# Boa sorte ;)

lista_pagamentos = []
lista_nao_pagos = []
for pgto in lista_de_pessoas:
    if pgto["isPay"] == True:
        lista_pagamentos.append(pgto)
    else:
        lista_nao_pagos.append(pgto)

print(f"Pagou: {lista_pagamentos}")
print(f"NAO Pagou: {lista_nao_pagos}")

pago_dinheiro = []
pago_cartao = []

for pag in lista_pagamentos:
    if pag["method"] == "Dinheiro":
        pago_dinheiro.append(pag)
    if pag["method"] == "Cartao":
        pago_cartao.append(pag)

print(f"Pagou no dinheiro : {pago_dinheiro}")
print(f"Pagou no cartao : {pago_cartao}")


```

### Exercicio 4

Em uma escola eu quero saber quem passou de ano. a media é 5.
Eu tambem preciso saber qual foi a Media dos alunos tanto aprovados quanto reprovados.

```python
lista_de_pessoas = [
    { "name": "Caio", "nota": 7.2 },
    { "name": "Clebinho", "nota": 4.5 },
    { "name": "Fulano", "nota": 9.4 },
    { "name": "Ciclano", "nota": 10.0 },
    { "name": "Joaozinho", "nota": 3.1 },
]

# Resultado
# Caio Fulano e Ciclano VÃO passaram de ano
# Clebinho e Joãozinho NÃO passaram de ano
# Media dos alunos aprovados: 8.87
# Media dos alunos aprovados: 3.80

# Boa sorte ;)

media = 5
aprovados = []
aprovados_notas = []
reprovados = []
reprovados_notas = []

for notas in lista_de_pessoas:
    if notas["nota"] >= media:
        aprovados.append(notas)
        aprovados_notas.append(notas['nota'])
    else:
        reprovados.append(notas)
        reprovados_notas.append(notas["nota"])

soma_aprovados = sum(aprovados_notas) / len(aprovados_notas)
soma_reprovados = sum(reprovados_notas) / len(reprovados_notas)


print(f"Os aprovados foram: {aprovados}")
print(f"Os reprovados foram: {reprovados}")
print(f'Media aprovados:{soma_aprovados:.2f}')
print(f'Media aprovados:{soma_reprovados:.2f}')
```

### Exercicio 5

No Exercito foi feito uma prova de resistencia onde os soldados tinham que ficar de baixo da agua. <br />
O Capitão quer saber a media geral dos soldados <br />
Tambem quer saber quem foi aprovado no teste quem ficou vai ter a chande de repetir e quem foi reprovado direto. <br />

Aprovados Media: 50 segundos <br />
Recuperação: 30 segundos <br />
Reprovados: abaixo de 30 segundos

```python
lista_de_soldados = [
    { "name": "Caio", "time": 0.52 },
    { "name": "Clebinho", "time": 0.34 },
    { "name": "Fulano", "time": .4 },
    { "name": "Ciclano", "time": 0.14 },
    { "name": "Joaozinho", "time": 0.29 },
]

# Resultado
# Media dos soldados: 0.34
# Caio PASSOU na seleção
# Clebinho e Fulano FICARAO de recuperão
# Ciclano e Joaozinho FORAM reprovados

# Boa sorte ;)


media = 0.50
media_recuperacao = 0.30
aprovado = []
recuperação = []
reprovado = []
somar_valores = []

for item in lista_de_soldados:
    somar_valores.append(item['time'])

    if item["time"] >= media:
        aprovado.append(item)
    
    if  item["time"] >= media_recuperacao and item["time"] <= media:
        recuperação.append(item)
    
    if item["time"] < 0.30:
        reprovado.append(item)


media_geral = sum(somar_valores) / len(somar_valores)

print(f"Os aprovados foram: {aprovado}")
print(f"Os recupeção: {recuperação}")
print(f"Os reprovados: {reprovado}")
print(f"A media é: {media_geral:.2f}")

```

### Exercicio 6

Jurados de uma corrida precisa saber quais carros ficarão nas primeiras posições.
Eles querem saber quais foram os 3 primeiros carros mais rapidos!

Alem disso eles querem saber a media dos 3 primeiros carro e a media do restante!

Eles precisam saber na ordem!

```python
lista_de_carros = [
    { "id": 1, "time": 1.11 },
    { "id": 2, "time": 1.3 },
    { "id": 3, "time": .59 },
    { "id": 4, "time": 1.09 },
    { "id": 5, "time": 1.02 },
    { "id": 6, "time": 1.32 },
    { "id": 7, "time": 0.58 },
    { "id": 8, "time": 1.2 },
    { "id": 9, "time": 1.01 },
    { "id": 10, "time": 1.29 },
]

# Resultado
# 3 Primeiros colocados: IDS 7, 3, 9
# Restante: IDS 5, 4, 1, 8, 10, 2, 6
# Media Primeiros colocados: 0.73
# Media Ultimos colocados: 1.19

# Boa sorte ;)

ordenado_tempo = sorted(lista_de_carros, key=lambda x: x["time"])
print(ordenado_tempo)

tres_primeiros = ordenado_tempo[:3]
print(f"Do início ao 3:   {tres_primeiros}")

restante = ordenado_tempo[4:]
print(f"Do 4 ao fim:   {restante}")


tempo_tres_primeiros = []

for tempo in tres_primeiros:
    tempo_tres_primeiros.append(tempo['time'])

print(tempo_tres_primeiros)

media = sum(tempo_tres_primeiros) / len(tempo_tres_primeiros)
print(f"A media é: {media:.2f}")

ordem_restante = []

for tempo in restante:
    ordem_restante.append(tempo['time'])

print(ordem_restante)

media1 = sum(ordem_restante) / len(ordem_restante)
print(f"A media é: {media1:.2f}")

```
