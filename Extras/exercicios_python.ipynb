### Exercicio 8.10 - Validador de CPF

Peca um CPF (11 digitos) e valide usando o algoritmo oficial (2 digitos verificadores).

````python
# SUA SOLUCAO AQUI:
````

<details><summary><b>Gabarito 8.10</b> (clique para revelar)</summary>

```python
cpf = input("CPF (11 digitos): ")
if len(cpf) != 11 or not cpf.isdigit():
    print("CPF invalido!")
else:
    # Primeiro digito
    soma = 0
    for i in range(9):
        soma += int(cpf[i]) * (10 - i)
    d1 = 11 - (soma % 11)
    if d1 >= 10:
        d1 = 0
    # Segundo digito
    soma = 0
    for i in range(10):
        soma += int(cpf[i]) * (11 - i)
    d2 = 11 - (soma % 11)
    if d2 >= 10:
        d2 = 0
    if int(cpf[9]) == d1 and int(cpf[10]) == d2:
        print("CPF valido!")
    else:
        print("CPF invalido!")
```
