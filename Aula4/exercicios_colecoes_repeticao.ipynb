### Exercício 2 — Soma e média
Dada a lista `notas = [7.5, 8.0, 6.5, 9.0, 7.0]`, calcule e exiba a soma e a média das notas usando um `for`.

**Saída esperada:** `Soma: 38.0` e `Média: 7.6`

````python
# Escreva seu código aqui
notas = [7.5, 8.0, 6.5, 9.0, 7.0]
soma = 0

for num in notas:
    soma += num

media = soma / len(notas)
print(f"{notas}: total {soma}→ média {media:.1f}")
````

### Exercício 50 — Boletim escolar completo
Dadas as listas:
```python
alunos = ["Ana", "Bruno", "Carlos", "Diana", "Eduardo"]
notas_p1 = [8.0, 6.5, 9.0, 5.0, 7.5]
notas_p2 = [7.5, 7.0, 8.5, 6.0, 8.0]
```
Calcule a média de cada aluno `(P1 + P2) / 2`, exiba um boletim formatado com nome, média e situação (Aprovado se média >= 7, senão Reprovado), e ao final exiba:
- Quantidade de aprovados e reprovados
- A maior e a menor média da turma

````python
# Escreva seu código aqui
#transformar em dicionario
#media de cada aluno
#aprovado / reprovado
#mostrar boletim
#quantidade aprovados e reprovados
#mair e menor media

alunos = ["Ana", "Bruno", "Carlos", "Diana", "Eduardo"]
notas_p1 = [8.0, 6.5, 9.0, 5.0, 7.5]
notas_p2 = [7.5, 7.0, 8.5, 6.0, 8.0]
boletim_final = []
aprovados, reprovados, maior, menor = 0, 0, None, None
#transformar em dicionario

for i in range(len(alunos)):
    dicio = {"nome": alunos[i],"P1":notas_p1[i], "P2":notas_p2[i]}
    dicio["media"] = (notas_p1[i] + notas_p2[i]) / 2  
    if (dicio["media"] >= 7 ):
        dicio["situacao"] = " APROVADO"
        aprovados +=1
    else:
        dicio["situacao"] = " REPROVADO"
        reprovados += 1
    boletim_final.append(dicio)

maior = boletim_final[0]["media"]
menor = boletim_final[0]["media"]
for aluno in boletim_final:
    if maior > aluno["media"]:
        maior = aluno["media"]
    if menor > aluno["media"]:
        menor = aluno["media"]

print('=' * 50)
for aluno in boletim_final:
    print("Nome: ", aluno['nome'], ' \nMedia: ' + str(aluno['media']) + ' \nSituação:' + aluno['situacao'])
    print('=' * 50)
    print("Aprovados", aprovados, "Reprovado", reprovados)
````
