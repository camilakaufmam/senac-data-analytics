## 📝 **PARTE 1: Cálculo Manual (SEM BIBLIOTECAS)**

```python
# ========================================
# PARTE 1: CÁLCULO MANUAL COMPLETO
# ========================================

# Dados dos vendedores
joao = [5000, 4800, 5200, 5100, 4900, 5000]
maria = [2000, 8000, 1500, 9500, 3000, 6000]

print("=" * 70)
print("PARTE 1: CÁLCULO MANUAL - VENDEDOR JOÃO")
print("=" * 70)

# 1. Média
media_joao = sum(joao) / len(joao)
print(f"\n1️⃣ Média: R$ {media_joao:.2f}")
print(f"   Cálculo: ({' + '.join(map(str, joao))}) / {len(joao)}")
print(f"   Cálculo: {sum(joao)} / {len(joao)} = {media_joao:.2f}")

# 2. Desvios (diferenças em relação à média)
desvios_joao = [x - media_joao for x in joao]
print(f"\n2️⃣ Desvios (x - média):")
for i, (valor, desvio) in enumerate(zip(joao, desvios_joao), 1):
    print(f"   Mês {i}: {valor} - {media_joao:.2f} = {desvio:.2f}")
print(f"   Lista de desvios: {[round(d, 2) for d in desvios_joao]}")

# 3. Quadrados dos desvios
quadrados_joao = [d ** 2 for d in desvios_joao]
print(f"\n3️⃣ Quadrados dos desvios (d²):")
for desvio, quadrado in zip(desvios_joao, quadrados_joao):
    print(f"   ({desvio:.2f})² = {quadrado:.2f}")
print(f"   Lista de quadrados: {[round(q, 2) for q in quadrados_joao]}")

# 4. Variância
variancia_joao = sum(quadrados_joao) / len(quadrados_joao)
print(f"\n4️⃣ Variância (média dos quadrados):")
print(f"   Cálculo: {sum(quadrados_joao):.2f} / {len(quadrados_joao)}")
print(f"   Variância = {variancia_joao:.2f}")

# 5. Desvio Padrão
desvio_padrao_joao = variancia_joao ** 0.5
print(f"\n5️⃣ Desvio Padrão (√variância):")
print(f"   Cálculo: √{variancia_joao:.2f}")
print(f"   Desvio Padrão = R$ {desvio_padrao_joao:.2f}")

# 6. Coeficiente de Variação
cv_joao = (desvio_padrao_joao / media_joao) * 100
print(f"\n6️⃣ Coeficiente de Variação (CV):")
print(f"   Cálculo: ({desvio_padrao_joao:.2f} / {media_joao:.2f}) × 100")
print(f"   CV = {cv_joao:.2f}%")

print("\n" + "=" * 70)
print("PARTE 1: CÁLCULO MANUAL - VENDEDORA MARIA")
print("=" * 70)

# 1. Média
media_maria = sum(maria) / len(maria)
print(f"\n1️⃣ Média: R$ {media_maria:.2f}")
print(f"   Cálculo: ({' + '.join(map(str, maria))}) / {len(maria)}")
print(f"   Cálculo: {sum(maria)} / {len(maria)} = {media_maria:.2f}")

# 2. Desvios
desvios_maria = [x - media_maria for x in maria]
print(f"\n2️⃣ Desvios (x - média):")
for i, (valor, desvio) in enumerate(zip(maria, desvios_maria), 1):
    print(f"   Mês {i}: {valor} - {media_maria:.2f} = {desvio:.2f}")
print(f"   Lista de desvios: {[round(d, 2) for d in desvios_maria]}")

# 3. Quadrados dos desvios
quadrados_maria = [d ** 2 for d in desvios_maria]
print(f"\n3️⃣ Quadrados dos desvios (d²):")
for desvio, quadrado in zip(desvios_maria, quadrados_maria):
    print(f"   ({desvio:.2f})² = {quadrado:.2f}")
print(f"   Lista de quadrados: {[round(q, 2) for q in quadrados_maria]}")

# 4. Variância
variancia_maria = sum(quadrados_maria) / len(quadrados_maria)
print(f"\n4️⃣ Variância (média dos quadrados):")
print(f"   Cálculo: {sum(quadrados_maria):.2f} / {len(quadrados_maria)}")
print(f"   Variância = {variancia_maria:.2f}")

# 5. Desvio Padrão
desvio_padrao_maria = variancia_maria ** 0.5
print(f"\n5️⃣ Desvio Padrão (√variância):")
print(f"   Cálculo: √{variancia_maria:.2f}")
print(f"   Desvio Padrão = R$ {desvio_padrao_maria:.2f}")

# 6. Coeficiente de Variação
cv_maria = (desvio_padrao_maria / media_maria) * 100
print(f"\n6️⃣ Coeficiente de Variação (CV):")
print(f"   Cálculo: ({desvio_padrao_maria:.2f} / {media_maria:.2f}) × 100")
print(f"   CV = {cv_maria:.2f}%")

print("\n" + "=" * 70)
print("✅ CÁLCULOS MANUAIS CONCLUÍDOS")
print("=" * 70)
```

## 📊 **PARTE 2: Validação com Bibliotecas**

```python
# ========================================
# PARTE 2: VALIDAÇÃO COM PANDAS E NUMPY
# ========================================

import pandas as pd
import numpy as np

# Dados dos vendedores
joao = [5000, 4800, 5200, 5100, 4900, 5000]
maria = [2000, 8000, 1500, 9500, 3000, 6000]

# Criar DataFrame
df = pd.DataFrame({
    'João': joao,
    'Maria': maria
})

print("=" * 70)
print("PARTE 2: VALIDAÇÃO COM BIBLIOTECAS")
print("=" * 70)

print("\n📊 DataFrame criado:")
print(df)

# Médias
print("\n" + "=" * 70)
print("1️⃣ MÉDIAS")
print("=" * 70)
print(df.mean())

# Variância (usando ddof=0 para variância populacional, como no cálculo manual)
print("\n" + "=" * 70)
print("2️⃣ VARIÂNCIA (populacional, N)")
print("=" * 70)
print(df.var(ddof=0))

# Desvio Padrão (usando ddof=0)
print("\n" + "=" * 70)
print("3️⃣ DESVIO PADRÃO (populacional, N)")
print("=" * 70)
print(df.std(ddof=0))

# Coeficiente de Variação
print("\n" + "=" * 70)
print("4️⃣ COEFICIENTE DE VARIAÇÃO (CV)")
print("=" * 70)
cv_joao_pd = (df['João'].std(ddof=0) / df['João'].mean()) * 100
cv_maria_pd = (df['Maria'].std(ddof=0) / df['Maria'].mean()) * 100

print(f"João:  {cv_joao_pd:.2f}%")
print(f"Maria: {cv_maria_pd:.2f}%")

# Resumo estatístico completo
print("\n" + "=" * 70)
print("5️⃣ RESUMO ESTATÍSTICO COMPLETO (.describe())")
print("=" * 70)
print(df.describe())

# Comparação com resultados manuais
print("\n" + "=" * 70)
print("✅ COMPARAÇÃO: Manual vs Pandas")
print("=" * 70)
print(f"\nJoão:")
print(f"  Média       - Manual: R$ {media_joao:.2f} | Pandas: R$ {df['João'].mean():.2f}")
print(f"  Variância   - Manual: {variancia_joao:.2f} | Pandas: {df['João'].var(ddof=0):.2f}")
print(f"  Desvio Pad. - Manual: R$ {desvio_padrao_joao:.2f} | Pandas: R$ {df['João'].std(ddof=0):.2f}")
print(f"  CV          - Manual: {cv_joao:.2f}% | Pandas: {cv_joao_pd:.2f}%")

print(f"\nMaria:")
print(f"  Média       - Manual: R$ {media_maria:.2f} | Pandas: R$ {df['Maria'].mean():.2f}")
print(f"  Variância   - Manual: {variancia_maria:.2f} | Pandas: {df['Maria'].var(ddof=0):.2f}")
print(f"  Desvio Pad. - Manual: R$ {desvio_padrao_maria:.2f} | Pandas: R$ {df['Maria'].std(ddof=0):.2f}")
print(f"  CV          - Manual: {cv_maria:.2f}% | Pandas: {cv_maria_pd:.2f}%")

print("\n✅ Os resultados são idênticos! Cálculo manual está correto.")
```

## 🏆 **PARTE 3: Análise e Decisão**

### **1. Quem tem a média de faturamento maior?**

**Resposta:** Ambos têm a **mesma média de faturamento**: R$ 5.000,00 por mês.

- João: R$ 5.000,00
- Maria: R$ 5.000,00

**Observação importante:** O fato de terem a mesma média é exatamente o que torna este problema interessante! A média sozinha **não revela** a diferença fundamental entre os dois vendedores.

### **2. Quem tem o menor Desvio Padrão? O que isso significa na prática?**

**Resposta:** **João** tem o menor desvio padrão.

- João: R$ 141,42 
- Maria: R$ 2.697,22

**O que isso significa:**

O desvio padrão mede o quanto os valores se afastam da média. Um desvio padrão **baixo** indica que os dados estão **concentrados próximos à média** (mais previsíveis), enquanto um desvio padrão **alto** indica que os dados estão **espalhados** (mais imprevisíveis).

**Na prática:**
- **João:** Suas vendas variam muito pouco (apenas ±R$ 141 da média). Isso significa que ele é **consistente e previsível**. A empresa pode confiar que ele sempre venderá próximo de R$ 5.000.
  
- **Maria:** Suas vendas variam muito (±R$ 2.697 da média). Pode ter meses excelentes (R$ 9.500) mas também meses muito ruins (R$ 1.500). Isso dificulta o **planejamento** da empresa.

### **3. Compare os Coeficientes de Variação (CV):**

- **CV do João:** 2,83%
- **CV da Maria:** 53,94%

**Classificação:**

| Vendedor | CV | Classificação | Interpretação |
|----------|-----|---------------|---------------|
| **João** | 2,83% | **Baixa dispersão** (< 15%) | Dados muito homogêneos, extremamente previsível |
| **Maria** | 53,94% | **Alta dispersão** (> 30%) | Dados muito heterogêneos, altamente imprevisível |

**Análise:**
- João tem um CV **extremamente baixo** (2,83%), indicando que sua variação representa apenas 2,83% da média. Ele é **excepcionalmente consistente**.
- Maria tem um CV **muito alto** (53,94%), indicando que sua variação representa mais da metade da média. Seu desempenho é **altamente volátil**.

### **4. DECISÃO FINAL: Quem merece o prêmio de 'Funcionário de Confiança'?**

# 🏆 **VENCEDOR: JOÃO**

---

### **Justificativa Completa:**

#### **Por que a média sozinha não é suficiente?**

A média aritmética nos diz apenas o valor **central** dos dados, mas não revela nada sobre a **variabilidade**. No caso de João e Maria:

- Ambos têm média de R$ 5.000,00
- Mas João varia entre R$ 4.800 e R$ 5.200 (amplitude de R$ 400)
- Enquanto Maria varia entre R$ 1.500 e R$ 9.500 (amplitude de R$ 8.000!)

Confiar apenas na média seria como dizer que "em média, se você colocar um pé no gelo e outro no fogo, a temperatura estará confortável" — a média esconde informações cruciais!

#### **O que o Coeficiente de Variação revela?**

O CV transforma a dispersão em um **percentual relativo à média**, permitindo avaliar o quão "confiável" é aquela média:

- **João (CV = 2,83%):** A variação é mínima em relação à média. Podemos **confiar** que ele entregará consistentemente próximo de R$ 5.000.

- **Maria (CV = 53,94%):** A variação é maior que **metade da média**. Não podemos confiar em previsões baseadas nessa média, pois o resultado real pode ser muito diferente.

#### **Por que a estabilidade/previsibilidade é importante?**

No contexto empresarial, a **previsibilidade** é fundamental para:

1. **Planejamento Financeiro:** A empresa precisa prever receitas para planejar investimentos, pagamentos, contratações.

2. **Gestão de Estoque:** Vendas previsíveis permitem otimizar estoques sem excessos ou faltas.

3. **Confiabilidade para Metas:** João pode receber metas mais desafiadoras com confiança de que as cumprirá. Maria pode decepcionar mesmo com potencial alto.

4. **Gestão de Risco:** Baixa variabilidade = baixo risco. A empresa pode contar com João nos momentos críticos.

5. **Exemplo para a Equipe:** Um funcionário consistente serve de referência para boas práticas e processos replicáveis.

#### **Observação sobre Maria:**

Embora Maria tenha meses excepcionais (R$ 9.500 em abril), ela também tem meses muito ruins (R$ 1.500 em março). Isso pode indicar:
- Dependência de grandes negócios esporádicos
- Falta de processo consistente
- Possível necessidade de treinamento ou suporte

Maria tem **potencial** (mostra que consegue vender muito), mas precisa desenvolver **consistência**.

---

### **Conclusão:**

João merece o prêmio de **'Funcionário de Confiança'** porque demonstra o atributo mais valioso no contexto empresarial: **previsibilidade consistente**. Sua baixíssima variância (CV = 2,83%) indica processos bem estabelecidos, disciplina e confiabilidade — exatamente o que uma empresa de sucesso busca em seus colaboradores-chave.

## 🎨 **BÔNUS: Visualizações**

```python
# ========================================
# VISUALIZAÇÕES COMPLETAS
# ========================================

import matplotlib.pyplot as plt
import seaborn as sns

# Dados dos vendedores
joao = [5000, 4800, 5200, 5100, 4900, 5000]
maria = [2000, 8000, 1500, 9500, 3000, 6000]
meses = ['Jan', 'Fev', 'Mar', 'Abr', 'Mai', 'Jun']

# Configurar estilo
sns.set_style("whitegrid")
plt.rcParams['figure.figsize'] = (16, 10)

# Criar figura com múltiplos gráficos
fig = plt.figure()

# Gráfico 1: Evolução Mensal (Linha)
plt.subplot(2, 2, 1)
plt.plot(meses, joao, 'go-', linewidth=2.5, markersize=10, label='João (Estável)', alpha=0.8)
plt.plot(meses, maria, 'ro-', linewidth=2.5, markersize=10, label='Maria (Instável)', alpha=0.8)
plt.axhline(5000, color='gray', linestyle='--', linewidth=1.5, alpha=0.6, label='Média (R$ 5.000)')
plt.title('Evolução do Faturamento Mensal', fontsize=14, fontweight='bold')
plt.ylabel('Faturamento (R$)', fontsize=11)
plt.xlabel('Mês', fontsize=11)
plt.legend(loc='best', fontsize=10)
plt.grid(True, alpha=0.3)
plt.ylim(0, 10000)

# Gráfico 2: Boxplot
plt.subplot(2, 2, 2)
df = pd.DataFrame({'João': joao, 'Maria': maria})
sns.boxplot(data=df, palette=['lightgreen', 'lightcoral'])
plt.title('Análise de Dispersão (Boxplot)', fontsize=14, fontweight='bold')
plt.ylabel('Faturamento (R$)', fontsize=11)
plt.grid(axis='y', alpha=0.3)

# Gráfico 3: Barras com Desvio Padrão
plt.subplot(2, 2, 3)
medias = [np.mean(joao), np.mean(maria)]
desvios = [np.std(joao, ddof=0), np.std(maria, ddof=0)]
vendedores = ['João', 'Maria']
cores = ['green', 'red']

bars = plt.bar(vendedores, medias, yerr=desvios, capsize=10, color=cores, alpha=0.7, edgecolor='black', linewidth=1.5)
plt.title('Média e Desvio Padrão', fontsize=14, fontweight='bold')
plt.ylabel('Faturamento Médio (R$)', fontsize=11)
plt.ylim(0, 8000)
plt.grid(axis='y', alpha=0.3)

# Adicionar valores nas barras
for i, (bar, media, desvio) in enumerate(zip(bars, medias, desvios)):
    plt.text(bar.get_x() + bar.get_width()/2, media + desvio + 200, 
             f'R$ {media:.0f}\n±{desvio:.0f}', 
             ha='center', va='bottom', fontsize=10, fontweight='bold')

# Gráfico 4: Coeficiente de Variação
plt.subplot(2, 2, 4)
cvs = [cv_joao, cv_maria]
cores_cv = ['darkgreen', 'darkred']
bars_cv = plt.barh(vendedores, cvs, color=cores_cv, alpha=0.7, edgecolor='black', linewidth=1.5)

# Adicionar linhas de referência
plt.axvline(15, color='orange', linestyle='--', linewidth=2, alpha=0.6, label='Limite Baixa Dispersão (15%)')
plt.axvline(30, color='red', linestyle='--', linewidth=2, alpha=0.6, label='Limite Alta Dispersão (30%)')

plt.title('Coeficiente de Variação (CV)', fontsize=14, fontweight='bold')
plt.xlabel('CV (%)', fontsize=11)
plt.legend(loc='best', fontsize=9)
plt.grid(axis='x', alpha=0.3)

# Adicionar valores nas barras
for i, (bar, cv) in enumerate(zip(bars_cv, cvs)):
    plt.text(cv + 1, bar.get_y() + bar.get_height()/2, 
             f'{cv:.2f}%', 
             va='center', fontsize=11, fontweight='bold')

plt.tight_layout()
plt.suptitle('Análise Comparativa: João vs Maria', fontsize=16, fontweight='bold', y=1.02)
plt.show()

print("\n✅ Visualizações geradas com sucesso!")
```
