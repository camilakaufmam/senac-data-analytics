## Explorando Variância, Desvio Padrão e Coeficiente de Variação

**Objetivo:** Compreender a variabilidade de um conjunto de dados e aprender a medir o quão "confiável" é uma média aritmética utilizando Python.

### 1. O Problema da Média: Um Exemplo Comparativo
Imagine duas rotas de entrega. Ambas têm o mesmo tempo médio de entrega (10 dias). Qual você escolheria?

* **Rota A:** [10, 10, 10, 10, 10]
* **Rota B:** [2, 18, 5, 15, 10]

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

rota_a = [10, 10, 10, 10, 10]
rota_b = [2, 18, 5, 15, 10]

print(f"Média Rota A: {np.mean(rota_a)}")
print(f"Média Rota B: {np.mean(rota_b)}")

# Visualização da diferença
plt.figure(figsize=(10, 4))
plt.plot(rota_a, 'go-', label='Rota A (Estável)')
plt.plot(rota_b, 'ro-', label='Rota B (Instável)')
plt.axhline(10, color='gray', linestyle='--')
plt.title("Mesma Média, Comportamentos Diferentes")
plt.legend()
plt.show()
```

### 2. Cálculo Manual: A Lógica por trás dos Números
Para medir essa "instabilidade", seguimos 5 passos matemáticos.

```python
# Vamos usar os dados da Rota B para o cálculo
dados = [2, 18, 5, 15, 10]

# Passo 1: Calcular a média
media = sum(dados) / len(dados)
print(f"1. Média: {media}")

# Passo 2: Diferenças em relação à média (O quanto cada ponto errou o alvo)
desvios = [x - media for x in dados]
print(f"2. Desvios: {desvios}")

# Passo 3: Elevar ao quadrado (para eliminar sinais negativos e dar peso aos erros maiores)
quadrados = [x**2 for x in desvios]
print(f"3. Quadrados dos desvios: {quadrados}")

# Passo 4: Variância (Média dos erros quadráticos)
variancia = sum(quadrados) / len(quadrados)
print(f"4. Variância: {variancia:.2f}")

# Passo 5: Desvio Padrão (Voltando para a unidade original dos dados)
desvio_padrao = variancia ** 0.5
print(f"5. Desvio Padrão: {desvio_padrao:.2f} dias")
```

### 3. Automatizando com Pandas e Numpy
No dia a dia do analista, usamos funções prontas que lidam com milhões de linhas instantaneamente.

```python
df = pd.DataFrame({'Rota_B': [2, 18, 5, 15, 10]})

var_np = np.var(df['Rota_B']) # Variância Populacional
std_pd = df['Rota_B'].std()   # Desvio Padrão Amostral (Padrão do mercado)

print(f"Variância (Numpy): {var_np:.2f}")
print(f"Desvio Padrão (Pandas): {std_pd:.2f}")

# O método .describe() resume tudo!
print("\nResumo Estatístico Completo:")
print(df.describe())
```

### 4. Interpretando os Resultados (Métricas de Confiança)
Como saber se a dispersão é aceitável? Usamos a métrica de **Distância** e o **CV**.

```python
# Cálculo da Distância (Variância / Média²)
distancia = variancia / (media ** 2)
print(f"Distância Relativa: {distancia:.2%}")

if distancia <= 0.10:
    status = "Baixa dispersão (Média MUITO confiável)"
elif distancia < 0.25:
    status = "Dispersão moderada (Média aceitável)"
else:
    status = "ALTA DISPERSÃO (Média NÃO confiável para previsões)"

print(f"Status: {status}")

# Coeficiente de Variação (Desvio Padrão / Média)
cv = (desvio_padrao / media) * 100
print(f"Coeficiente de Variação (CV): {cv:.2f}%")
```

### 5. Visualização Profissional: Boxplot
O Boxplot é a ferramenta definitiva para ver a dispersão e identificar *Outliers* (valores fora da curva).

```python
# Gerando dados sintéticos para simular 100 entregas
np.random.seed(42)
entregas_estaveis = np.random.normal(10, 1, 100)
entregas_instaveis = np.random.normal(10, 5, 100)

plt.figure(figsize=(12, 6))
sns.boxplot(data=[entregas_estaveis, entregas_instaveis], palette="Set2")
plt.xticks([0, 1], ['Fábrica A (Estável)', 'Fábrica B (Instável)'])
plt.title("Análise de Dispersão das Fábricas")
plt.grid(axis='y', alpha=0.3)
plt.show()
```

```python
# ============================================================
# GABARITO COMPLETO - SOMENTE PARA O PROFESSOR
# ============================================================

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

joao = [5000, 4800, 5200, 5100, 4900, 5000]
maria = [2000, 8000, 1500, 9500, 3000, 6000]

# ========================================
# PARTE 1: CÁLCULO MANUAL
# ========================================

print("=" * 60)
print("PARTE 1: CÁLCULO MANUAL")
print("=" * 60)

# JOÃO
print("\n📊 VENDEDOR JOÃO:")
media_joao = sum(joao) / len(joao)
print(f"1. Média: R$ {media_joao:.2f}")

desvios_joao = [x - media_joao for x in joao]
print(f"2. Desvios: {[round(d, 2) for d in desvios_joao]}")

quadrados_joao = [d**2 for d in desvios_joao]
print(f"3. Quadrados: {[round(q, 2) for q in quadrados_joao]}")

variancia_joao = sum(quadrados_joao) / len(quadrados_joao)
print(f"4. Variância: {variancia_joao:.2f}")

desvio_padrao_joao = variancia_joao ** 0.5
print(f"5. Desvio Padrão: R$ {desvio_padrao_joao:.2f}")

cv_joao = (desvio_padrao_joao / media_joao) * 100
print(f"6. CV: {cv_joao:.2f}%")

# MARIA
print("\n📊 VENDEDOR MARIA:")
media_maria = sum(maria) / len(maria)
print(f"1. Média: R$ {media_maria:.2f}")

desvios_maria = [x - media_maria for x in maria]
print(f"2. Desvios: {[round(d, 2) for d in desvios_maria]}")

quadrados_maria = [d**2 for d in desvios_maria]
print(f"3. Quadrados: {[round(q, 2) for q in quadrados_maria]}")

variancia_maria = sum(quadrados_maria) / len(quadrados_maria)
print(f"4. Variância: {variancia_maria:.2f}")

desvio_padrao_maria = variancia_maria ** 0.5
print(f"5. Desvio Padrão: R$ {desvio_padrao_maria:.2f}")

cv_maria = (desvio_padrao_maria / media_maria) * 100
print(f"6. CV: {cv_maria:.2f}%")

# ========================================
# PARTE 2: COM BIBLIOTECAS
# ========================================

print("\n" + "=" * 60)
print("PARTE 2: VALIDAÇÃO COM BIBLIOTECAS")
print("=" * 60)

df = pd.DataFrame({
    'João': joao,
    'Maria': maria
})

print("\n📈 Estatísticas com Pandas:")
print(f"\nMédias:\n{df.mean()}")
print(f"\nDesvio Padrão:\n{df.std()}")
print(f"\nVariância:\n{df.var()}")

# Coeficientes de Variação
cv_joao_pd = (df['João'].std() / df['João'].mean()) * 100
cv_maria_pd = (df['Maria'].std() / df['Maria'].mean()) * 100

print(f"\nCoeficiente de Variação:")
print(f"João: {cv_joao_pd:.2f}%")
print(f"Maria: {cv_maria_pd:.2f}%")

print("\n📊 Resumo Estatístico Completo:")
print(df.describe())

# ========================================
# PARTE 3: ANÁLISE E DECISÃO
# ========================================

print("\n" + "=" * 60)
print("PARTE 3: ANÁLISE E DECISÃO")
print("=" * 60)

print("\n1️⃣ Média de faturamento:")
print(f"   Ambos têm a mesma média: R$ 5.000,00")
print(f"   ⚠️ A média SOZINHA não revela a diferença entre eles!")

print("\n2️⃣ Desvio Padrão:")
print(f"   João: R$ {desvio_padrao_joao:.2f} (MENOR)")
print(f"   Maria: R$ {desvio_padrao_maria:.2f} (MAIOR)")
print(f"   → João tem variação muito menor, mais previsível")

print("\n3️⃣ Coeficiente de Variação:")
print(f"   João: {cv_joao:.2f}% → BAIXA DISPERSÃO (< 15%)")
print(f"   Maria: {cv_maria:.2f}% → ALTA DISPERSÃO (> 30%)")

print("\n4️⃣ DECISÃO FINAL:")
print(f"   🏆 VENCEDOR: JOÃO")
print(f"\n   💡 JUSTIFICATIVA:")
print(f"   • Ambos faturam em média R$ 5.000/mês")
print(f"   • João tem CV de {cv_joao:.2f}%, indicando alta previsibilidade")
print(f"   • Maria tem CV de {cv_maria:.2f}%, indicando vendas muito voláteis")
print(f"   • Para gestão e planejamento, ESTABILIDADE é crucial")
print(f"   • João permite prever receita com confiança")
print(f"   • Maria pode ter meses excelentes, mas também péssimos")

# ========================================
# BÔNUS: VISUALIZAÇÕES
# ========================================

print("\n" + "=" * 60)
print("BÔNUS: VISUALIZAÇÕES")
print("=" * 60)

fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Gráfico 1: Evolução mensal
meses = ['Jan', 'Fev', 'Mar', 'Abr', 'Mai', 'Jun']
axes[0].plot(meses, joao, 'go-', linewidth=2, markersize=8, label='João (Estável)')
axes[0].plot(meses, maria, 'ro-', linewidth=2, markersize=8, label='Maria (Instável)')
axes[0].axhline(5000, color='gray', linestyle='--', alpha=0.5, label='Média')
axes[0].set_title('Evolução do Faturamento - Semestre', fontsize=12, fontweight='bold')
axes[0].set_ylabel('Faturamento (R$)')
axes[0].set_xlabel('Mês')
axes[0].legend()
axes[0].grid(True, alpha=0.3)

# Gráfico 2: Boxplot
sns.boxplot(data=df, ax=axes[1], palette=['lightgreen', 'lightcoral'])
axes[1].set_title('Análise de Dispersão', fontsize=12, fontweight='bold')
axes[1].set_ylabel('Faturamento (R$)')
axes[1].grid(axis='y', alpha=0.3)

plt.tight_layout()
plt.show()

print("\n✅ Análise completa finalizada!")
```
