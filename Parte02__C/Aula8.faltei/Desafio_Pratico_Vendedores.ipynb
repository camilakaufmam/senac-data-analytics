````python
# ========================================
# EXEMPLOS PRÁTICOS - EXECUTE E OBSERVE
# ========================================

# Exemplo 1: Laço for tradicional
print("=" * 50)
print("EXEMPLO 1: Laço FOR tradicional")
print("=" * 50)

frutas = ["maçã", "banana", "laranja"]
for fruta in frutas:
    print(f"Eu gosto de {fruta}")

# Exemplo 2: Criar nova lista com for
print("\n" + "=" * 50)
print("EXEMPLO 2: Criar lista com FOR")
print("=" * 50)

numeros = [1, 2, 3, 4, 5]
dobros = []
for num in numeros:
    dobros.append(num * 2)
    
print(f"Números originais: {numeros}")
print(f"Números dobrados: {dobros}")

# Exemplo 3: List Comprehension (mesma coisa, mais compacto)
print("\n" + "=" * 50)
print("EXEMPLO 3: List Comprehension")
print("=" * 50)

numeros = [1, 2, 3, 4, 5]
dobros = [num * 2 for num in numeros]

print(f"Números originais: {numeros}")
print(f"Números dobrados: {dobros}")

# Exemplo 4: Operações com variável externa
print("\n" + "=" * 50)
print("EXEMPLO 4: Usando variável externa")
print("=" * 50)

temperaturas = [28, 32, 30, 29]
temp_media = sum(temperaturas) / len(temperaturas)

# Calcular diferença de cada temperatura em relação à média
diferencas = [temp - temp_media for temp in temperaturas]

print(f"Temperaturas: {temperaturas}")
print(f"Média: {temp_media}°C")
print(f"Diferenças em relação à média: {diferencas}")

# Exemplo 5: Elevar ao quadrado
print("\n" + "=" * 50)
print("EXEMPLO 5: Elevar ao quadrado")
print("=" * 50)

valores = [2, -3, 4, -5]
quadrados = [v ** 2 for v in valores]

print(f"Valores originais: {valores}")
print(f"Valores ao quadrado: {quadrados}")
print("(Note que os negativos ficaram positivos!)")

# Exemplo 6: Raiz quadrada
print("\n" + "=" * 50)
print("EXEMPLO 6: Raiz quadrada")
print("=" * 50)

numeros = [4, 9, 16, 25]
raizes = [n ** 0.5 for n in numeros]

print(f"Números: {numeros}")
print(f"Raízes quadradas: {raizes}")

print("\n✅ Agora você entende laços de repetição!")
````

## 🔄 **Revisão: Laços de Repetição e List Comprehension**

Antes de iniciar o desafio, vamos revisar como **processar cada elemento de uma lista** usando laços de repetição — uma técnica fundamental para os cálculos estatísticos.

### **1. O que é um Laço de Repetição?**

Um **laço de repetição** permite executar a mesma operação para cada item de uma lista, sem precisar escrever código repetido.

**Problema:** Somar 10 a cada número de uma lista

```python
numeros = [5, 15, 25, 35]

# ❌ Forma manual (ruim, não escalável)
resultado1 = numeros[0] + 10
resultado2 = numeros[1] + 10
resultado3 = numeros[2] + 10
resultado4 = numeros[3] + 10

# ✅ Com laço for (eficiente)
resultados = []
for numero in numeros:
    novo_valor = numero + 10
    resultados.append(novo_valor)

print(resultados)  # [15, 25, 35, 45]
```

**Sintaxe do `for`:**
```python
for item in lista:
    # faça algo com item
```

---

### **2. List Comprehension: A Forma Pythônica**

**List comprehension** é uma forma compacta de criar novas listas a partir de listas existentes.

**Mesma operação, em UMA linha:**

```python
numeros = [5, 15, 25, 35]

# ✅ List Comprehension (RECOMENDADO)
resultados = [numero + 10 for numero in numeros]

print(resultados)  # [15, 25, 35, 45]
```

**Sintaxe:**
```python
nova_lista = [expressão for item in lista_original]
```

**Tradução em português:**
> "Crie uma nova lista aplicando a **expressão** para cada **item** da **lista_original**"

---

### **3. Exemplos Práticos**

#### **Exemplo A: Multiplicar cada elemento por 2**
```python
valores = [3, 7, 11]

# Laço tradicional
dobrados_for = []
for v in valores:
    dobrados_for.append(v * 2)

# List comprehension (mais limpo)
dobrados = [v * 2 for v in valores]

print(dobrados)  # [6, 14, 22]
```

#### **Exemplo B: Elevar cada elemento ao quadrado**
```python
numeros = [-3, 4, -5, 2]

# Elevar ao quadrado (elimina negativos)
quadrados = [n ** 2 for n in numeros]

print(quadrados)  # [9, 16, 25, 4]
```

#### **Exemplo C: Usando uma variável externa**
```python
precos = [100, 200, 150]
preco_medio = sum(precos) / len(precos)  # 150

# Calcular diferença de cada preço em relação à média
diferencas = [preco - preco_medio for preco in precos]

print(diferencas)  # [-50.0, 50.0, 0.0]
```
