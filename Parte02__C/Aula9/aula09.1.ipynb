```python
import pandas as pd

df = pd.read_csv('aula9.csv')
#print(df.head)

print("Analise Preço Vendas")
preco_venda = df['Preco_Venda']
#print(preco_venda)

print(f"Media: {preco_venda.mean():.2f}")
print(f"Mediana: {preco_venda.median():.2f}")
print(f"Devio Padrão: {preco_venda.std():.2f}")
```

```python
#Assimetria
assimetria_preco_venda = preco_venda.skew()
print(f"Assimetria (Preço de venda): {assimetria_preco_venda:.2f}")

#Interpretação da Assimetria
if -0.5 <= assimetria_preco_venda <= 0.5:
    print("Distribuição Simétrica (Média próxima da Mediana")
elif assimetria_preco_venda > 0.5:
    print("Distribuição Assimétrica Positiva (Média > Mediana)")
else:
    print("Distribuição Assimétrica Negativa (Média < Mediana)")    
```

```python
# CALCULAR A MEDIA, MEDIANA, DESVIO PADRÃO, ASSIMETRIA E INTERPRETAR A ASSIMETRIA DA COLUNA QUANTIDADE

print("Minha Analise de Preço da Coluna Quantidade")
quantidade = df['Quantidade']
#print(quantidade)

print(f"A média é: {quantidade.mean():.2f}")
print(f"A mediana é: {quantidade.median():.2f}")
print(f"O desvio padrão é: {quantidade.std():.2f}")

#ASSIMETRIA
assimetria_quantidade = quantidade.skew()
print(f"Assimetria da quantidade é: {assimetria_quantidade:.2f}")  #quando é so a variável, não precisa colocar os parenteses

#INTERPRETAÇÃO DA ASSIMETRIA
if -0.5 <= assimetria_quantidade <= 0.5:
    print("Distribuição Simétrica (Média próxima da Mediana)")
elif assimetria_quantidade > 0.5:
    print("Distribuição Assimétrica Positiva (Média > Mediana)")
else: 
    print("Distribuição Assimétrica Negativa (Média < Mediana)")
```

```python
#CURTOSE

curtose_preco_venda = preco_venda.kurtosis()
print(f"Curtose (Preço de Venda): {curtose_preco_venda:.2f}")

#Tipos de Curtose = Mesocútica, Platicúrtica e Leptocúrtica
if 2.5 <= curtose_preco_venda <= 3.5:
    print("Distribuição Mesocútica")
elif curtose_preco_venda < 2.5:
    print("Distribuição Platicúrtica")
else:
    print("Distribuição Leptocúrtica")
```

```python
curtose_quantidade = quantidade.kurtosis()
print(f"Curtose da Quantidade: {curtose_quantidade:.2f}")

#Tipos de Curtose = Mesocúrtica, Platicúrtica e Leptocúrtica
if 2.5 <= curtose_quantidade <= 3.5:
    print("Distribuição Mesocúrtica")
elif curtose_quantidade < 2.5:
    print("Ditribuição Platicúrtica")
else:
    print("Distribuição Leptocúrtica")
```
