```python
import mysql.connector
import pandas as pd

#Instale o pandas e mysql-connector-python usando pip:
# pip install pandas mysql-connector-python
# instale o openpyxl para ler arquivos Excel:
# pip install openpyxl


def obter_dados_do_banco(query):
    conn = None
    df = pd.DataFrame()
    try:
        # Configurações de conexão (ajuste conforme seu MySQL)
        conn = mysql.connector.connect(
            host="localhost",
            user="root",      # Usuário padrão do XAMPP/WAMP
            password="",      # Senha padrão do XAMPP/WAMP (geralmente vazia)
            port=3307,       # Porta padrão do MySQL - descomente se necessário
            database="Aula4"
        )
        
        if conn.is_connected():
            cursor = conn.cursor()
            cursor.execute(query)
            
            # Obter nomes das colunas
            column_names = [i[0] for i in cursor.description]
            
            # Obter todos os registros
            records = cursor.fetchall()
            
            # Criar DataFrame Pandas
            df = pd.DataFrame(records, columns=column_names)
            
    except mysql.connector.Error as err:
        print(f"Erro ao conectar ou executar query: {err}")
    finally:
        if conn and conn.is_connected():
            cursor.close()
            conn.close()
            print("Conexão com o MySQL encerrada.")
    return df


query_vendas_completas = """
SELECT
    v.ID_Venda,
    v.Data_Venda,
    v.Quantidade,
    p.ID_Produto, 
    p.Nome_Produto,
    p.Preco_Unitario,
    vend.ID_Vendedor, 
    vend.Nome_Vendedor,
    vend.Regiao
FROM vendas v
INNER JOIN produtos p ON v.ID_Produto = p.ID_Produto
INNER JOIN vendedores vend ON v.ID_Vendedor = vend.ID_Vendedor;"""

df_vendas_mysql = obter_dados_do_banco(query_vendas_completas)
print("\nDataFrame de Vendas do MySQL (head):\n", df_vendas_mysql.head()) 
```

```python
import numpy as np

# Array de vendas com outlier
vendas_produto_x = np.array([100, 120, 110, 130, 1000]) 
print("Array de Vendas: ", vendas_produto_x)

# Calcular Média e Mediana
media = np.mean(vendas_produto_x)
mediana = np.median(vendas_produto_x)

print(f"Média: {media:.2f}")
print(f"Mediana: {mediana:.2f}")

# Calcular a Distância
distancia = (media - mediana) / mediana
print(f"Distância (Média vs Mediana): {distancia:.2%}")  #2% = transforma em casa decimal e com 2 decimais

# Interpretação
if abs(distancia) <= 0.10:
    print("Interpretação: Distância <= |10%|. A média tende a ser uma medida de tendência central confiável.")
elif abs(distancia) < 0.25:
    print("Interpretação: Distância >= |10%| e < |25%|. A média está sofrendo moderadamente a influência dos valores extremos.")
else:
    print("Interpretação: Distância >= |25%|. A média está sofrendo alta influência dos valores extremos e tende a não ser uma medida de tendência central confiável.")

# Atenção:
if media > mediana:
    print("Atenção: A média é maior que a mediana, a influência é dos valores maiores da distribuição (outliers positivos).")
elif media < mediana:
    print("Atenção: A média é menor que a mediana, a influência é dos valores menores da distribuição (outliers negativos).")
else:
    print("Atenção: Média e mediana são iguais, indicando uma distribuição simétrica.")
```
