```python
import mysql.connector
import pandas as pd

#Instale o pandas e mysql-connector-python usando pip:
# pip install pandas mysql-connector-python
# instale o openpyxl para ler arquivos Excel:
# pip install openpyxl


def obter_dados_do_banco(query):
    conn = None
    df = pd.DataFrame()
    try:
        # Configurações de conexão (ajuste conforme seu MySQL)
        conn = mysql.connector.connect(
            host="localhost",
            user="root",      # Usuário padrão do XAMPP/WAMP
            password="",      # Senha padrão do XAMPP/WAMP (geralmente vazia)
            #port=3306,       # Porta padrão do MySQL - descomente se necessário
            database="dataorganized_relacional_db"
        )
        
        if conn.is_connected():
            cursor = conn.cursor()
            cursor.execute(query)
            
            # Obter nomes das colunas
            column_names = [i[0] for i in cursor.description]
            
            # Obter todos os registros
            records = cursor.fetchall()
            
            # Criar DataFrame Pandas
            df = pd.DataFrame(records, columns=column_names)
            
    except mysql.connector.Error as err:
        print(f"Erro ao conectar ou executar query: {err}")
    finally:
        if conn and conn.is_connected():
            cursor.close()
            conn.close()
            print("Conexão com o MySQL encerrada.")
    return df


query_vendas_completas = """
SELECT
    v.ID_Venda,
    v.Data_Venda,
    v.Quantidade,
    p.ID_Produto, 
    p.Nome_Produto,
    p.Preco_Unitario,
    vend.ID_Vendedor, 
    vend.Nome_Vendedor,
    vend.Regiao
FROM vendas v
INNER JOIN produtos p ON v.ID_Produto = p.ID_Produto
INNER JOIN vendedores vend ON v.ID_Vendedor = vend.ID_Vendedor;"""
```
