```python
#primeiro de tudo, declarar pd:
import pandas as pd

#importar CSV
df_vendas = pd.read_csv('vendas_aula7.csv')
df_produtos = pd.read_csv('produtos_aula7.csv')
#importar EXCEL
df_produtos_info = pd.read_excel('produtos_info.xlsx')

print(df_vendas.head())
print(df_produtos.head())
print(df_produtos_info.head())
```

```python
#MERGE (juntando duas planilhas com base em uma coluna em comum)
df_vendas_completo = pd.merge (df_vendas, df_produtos, on='ID_Produto', how='left') #left quer di
print(df_vendas_completo.head())

#left quer dizer que esta usando como base a planilha da esquerda (df_vendas)
#se fosse right, seria a planilha da direita (df_produtos)
```

```python
#MERGE 
df_final = pd.merge(df_vendas_completo,df_produtos_info, on='ID_Produto', how='left' )
print(df_final.head())
```

```python
#Crie a coluna de Faturamento

df_final ['Faturamento'] = df_final['Quantidade'] * df_final['Preco_Venda']
print(df_final.head())
#print(df_final.columns) para verificar as colunas
```

```python
            #Calcular Medidas Estatísticas para o Faturamento

#MEDIA:
media_faturamento = df_final["Faturamento"].mean()
print(f"Média do faturamento é: R$ {media_faturamento:.2f}")

#MEDIANA
mediana_faturamento = df_final["Faturamento"].median()
print(f"Mediana do faturamento é: R$ {mediana_faturamento:.2f}")

# Primeiro Quartil (Q1) e Segundo Quartil (Q3)

q1 = df_final['Faturamento'].quantile(0.25)
q3 = df_final['Faturamento'].quantile(0.75)

# Intervalo Interquartil (IQR = Q3 - Q1)
IQR = q3 - q1

print(f"O primeiro quartil é: {q1:.2f}")
print(f"O segundo quartil é: {q3:.2f}")
print(f"O intervalo interquartil é: {IQR:.2f}")

#Identificação de Outliers:

limite_inferior = q1 - (1.5 * IQR) 
print(limite_inferior)

limite_superior = q3 + (1.5 * IQR)
print(limite_superior)
```

```python
        #Medidas de Amplitude

#Menor Faturamento (mínimo):

menor_faturamento = df_final['Faturamento'].min()
print(f"O menor faturamento é : {menor_faturamento:.2f}")


#Maior Faturamento (máximo) 

maior_faturamento = df_final['Faturamento'].max()
print(f"O maior faturamento é: {maior_faturamento:.2f}")


#Amplitude Total = Máximo - Mínimo

amplitude_total = maior_faturamento - menor_faturamento
print(f"A amplitude total é: {amplitude_total:.2f}")
```
