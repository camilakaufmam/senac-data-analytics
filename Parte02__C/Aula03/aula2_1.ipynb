```python
dados1 = np.array([12, 15, 17, 20, 22, 25, 28, 30, 35, 40])

#primeiro quartil
q1 = np.percentile(dados1, 25)
#mediana
q2 = np.percentile(dados1, 50)
#segundo quartil
q3 = np.percentile(dados1, 75)

print(f"primeiro quartil (Q1): {q1}")
print(f"meidiana (Q2): {q2}")
print(f"segundo quartil(Q3): {q3}")
```

```python
import matplotlib.pyplot as plt

df_vendas["Categoria"].value_counts().plot(kind='pie', title='Vendas por Categoria')  #"kind"= para escolher o tipo de grafico;"bar"= é o grafico de Barra Vertical; 'barh' = Barra horizontal;
#pie = grafico de pizza
plt.ylabel("Número de Vendas") #nome do eixo y(em pé)
plt.xlabel("Categoria") #nome do eixo x(deitado)
plt.xticks(rotation=45) #nomes no eixo x em 45graus
plt.tight_layout()  #faz caber dentro do layout
plt.show()  #para mostrar
```
