```python
import pandas as pd

df_vendas = pd.read_excel("vendas_dataorganized.xlsx")

print(df_vendas.head()) #head para trazer o 5 primeiros itens
print(df_vendas.info())

faturamento_total = df_vendas["Faturamento_Total"].sum() #sum é a soma total
print(f"\nFaturamento Total: R$ {faturamento_total:.2f}")

media_vendas = df_vendas["Faturamento_Total"].mean()     #mean é a média
print(f"\nA media das vendas é : R$ {media_vendas:.2f}")

valor_maximo = df_vendas["Faturamento_Total"].max()     #max pega o valor maximo da tabela
print(f"\nO valor máximo de vendas é R$: {valor_maximo:.2f}")

valor_minimo = df_vendas["Faturamento_Total"].min()
print(f"\nO valor minimo é R$: {valor_minimo:.2f}")
```
