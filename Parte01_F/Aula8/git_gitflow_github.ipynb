## 1. Por que Controle de Versão? <a id="1"></a>

Imagine que você está desenvolvendo um projeto e salva arquivos assim:

```
projeto_final.py
projeto_final_v2.py
projeto_final_v2_CORRIGIDO.py
projeto_final_v2_CORRIGIDO_agora_vai.py
projeto_ESSE_AQUI.py
```

Isso é **controle de versão manual** — caótico, ineficiente e propenso a erros.

### O que um Sistema de Controle de Versão (VCS) resolve:

| Problema | Solução com Git |
|---|---|
| Perder código antigo | Histórico completo de cada alteração |
| Trabalho em equipe conflitante | Branches separadas por funcionalidade |
| Bug novo introduzido | Identificar exatamente qual commit causou |
| Rollback necessário | Voltar para qualquer versão anterior |
| Rastrear quem fez o quê | Cada commit tem autor, data e mensagem |

## 2. Git — Conceitos Fundamentais <a id="2"></a>

### O que é Git?

**Git** é um sistema de controle de versão **distribuído**, criado por **Linus Torvalds** em 2005 para gerenciar o desenvolvimento do kernel Linux.

**Distribuído** significa que cada desenvolvedor tem uma cópia **completa** do repositório — histórico incluso — na sua máquina.

### As Três Áreas do Git

```
┌───────────────────────────────────────────────────────────────┐
│                         MÁQUINA LOCAL                         │
│                                                               │
│  ┌──────────────┐    git add    ┌──────────────┐              │
│  │   WORKING    │ ──────────►  │   STAGING    │               │
│  │  DIRECTORY   │              │     AREA     │               │
│  │              │ ◄──────────  │  (Index)     │               │
│  │  Seus        │  git restore │              │               │
│  │  arquivos    │              └──────┬───────┘               │
│  └──────────────┘                     │ git commit            │
│                                       ▼                       │
│                              ┌──────────────┐                 │
│                              │  REPOSITORY  │                 │
│                              │   (.git/)    │                 │
│                              │  Histórico   │                 │
│                              │  completo    │                 │
│                              └──────────────┘                 │
└───────────────────────────────────────────────────────────────┘
```

### Estados de um Arquivo

```
  Não rastreado      Rastreado
  (Untracked)    ┌───────────────────────────────────┐
       │         │   Modified → Staged → Committed   │
       │         │                                   │
  ─────┼─────────┼──────────────────────────────────►│
       │  git add│                                   │
       ▼         └───────────────────────────────────┘
```

| Estado | Descrição |
|---|---|
| **Untracked** | Arquivo novo, Git ainda não conhece |
| **Modified** | Arquivo modificado desde o último commit |
| **Staged** | Marcado para entrar no próximo commit |
| **Committed** | Salvo permanentemente no repositório |

## 4. Repositório Local — Comandos Essenciais <a id="4"></a>

### 4.1 Iniciando um Repositório

```bash
# Criar novo projeto
mkdir meu-projeto
cd meu-projeto
git init
# Resultado: Initialized empty Git repository in .../meu-projeto/.git/

# OU clonar repositório existente
git clone https://github.com/usuario/repositorio.git
git clone https://github.com/usuario/repositorio.git nome-local
```

### 4.2 O Ciclo Diário de Trabalho

```bash
# 1. Ver o estado atual do projeto
git status

# 2. Ver diferenças antes de adicionar
git diff                     # diferenças no working directory
git diff --staged            # diferenças no staging area

# 3. Adicionar arquivos ao staging
git add arquivo.py           # arquivo específico
git add src/                 # pasta inteira
git add .                    # todos os arquivos modificados
git add *.py                 # por padrão de nome

# 4. Criar o commit
git commit -m "feat: adiciona função de login"
git commit                   # abre editor para mensagem longa

# Atalho: adicionar e commitar ao mesmo tempo (só arquivos já rastreados)
git commit -am "fix: corrige bug no cálculo de frete"
```

### 4.3 Inspecionando o Histórico

```bash
# Histórico completo
git log

# Histórico resumido (uma linha por commit)
git log --oneline

# Histórico com grafo de branches
git log --oneline --graph --all

# Histórico de um arquivo específico
git log --follow arquivo.py

# Buscar por autor
git log --author="João"

# Buscar por mensagem
git log --grep="login"

# Ver detalhes de um commit específico
git show abc1234
```

**Exemplo de saída do `git log --oneline --graph --all`:**
```
* 3f2a1c8 (HEAD -> main) feat: adiciona relatório de vendas
* 9b4e5d2 fix: corrige cálculo de desconto
* 7a8c3f1 feat: implementa carrinho de compras
* 2d6e4b0 docs: atualiza README
* 1a2b3c4 Initial commit
```

### 4.4 Desfazendo Alterações

```
SITUAÇÃO                   COMANDO
─────────────────────────────────────────────────────────────────
Arquivo modificado,        git restore arquivo.py
ainda não staged           (descarta mudanças no working dir)

Arquivo no staging,        git restore --staged arquivo.py
quer tirar do stage        (mantém as mudanças, só remove do stage)

Último commit errado,      git commit --amend -m "nova mensagem"
ainda não enviado          (reescreve o último commit)

Voltar N commits           git revert HEAD~N
mantendo histórico         (cria novo commit desfazendo as mudanças)

Resetar para commit        git reset --soft HEAD~1
(mantém staging)           (desfaz commit, mantém arquivos staged)

Resetar para commit        git reset --hard HEAD~1
(DESTRUTIVO!)              ⚠️ APAGA mudanças permanentemente
─────────────────────────────────────────────────────────────────
```

> **Regra de Ouro:** Nunca use `--hard` em commits já enviados para o repositório remoto!

### 4.5 O arquivo .gitignore

## 5. Branches — Ramificações do Projeto <a id="5"></a>

### O que é uma Branch?

Uma **branch** (ramificação) é uma linha de desenvolvimento independente. Ela permite trabalhar em funcionalidades, correções ou experimentos **sem afetar o código principal**.

```
main:     A ──── B ──── C ──────────────────── M
                  \                           /
feature:           D ──── E ──── F ──── G ──
                                (merge na main)
```

Cada letra representa um commit. A branch `feature` foi criada a partir de `B`, desenvolvida independentemente, e depois mesclada de volta à `main` no ponto `M`.

### 5.1 Comandos de Branch

```bash
# Listar branches locais
git branch

# Listar todas (locais + remotas)
git branch -a

# Criar nova branch
git branch feature/login

# Trocar para a branch
git checkout feature/login
# ou (forma moderna)
git switch feature/login

# Criar e já trocar (atalho)
git checkout -b feature/login
# ou
git switch -c feature/login

# Renomear branch atual
git branch -m novo-nome

# Deletar branch (após merge)
git branch -d feature/login

# Forçar deleção (sem merge)
git branch -D feature/login
```

### 5.2 Merge — Unindo Branches

```bash
# Estando na branch de destino (ex: main)
git checkout main

# Fazer o merge da feature
git merge feature/login
```

**Tipos de Merge:**

```
FAST-FORWARD MERGE (sem divergência):
─────────────────────────────────────
Antes:  main: A ── B
               \── C ── D  (feature)

Depois: main: A ── B ── C ── D  ✓ (limpo)


THREE-WAY MERGE (com divergência):
───────────────────────────────────
Antes:  main:    A ── B ── E
                  \── C ── D  (feature)

Depois: main: A ── B ── E ──┐
               \── C ── D ──┴── M (commit de merge)
```

### 5.3 Rebase — Histórico Linear

```bash
# Na branch feature, 'replanta' sobre a main
git checkout feature/login
git rebase main
```

```
REBASE:
─────────────────────────────────────────────────────
Antes:  main:    A ── B ── E
                  \── C ── D  (feature)

Depois: main:    A ── B ── E ── C' ── D'  (histórico linear)
```

| | Merge | Rebase |
|---|---|---|
| **Histórico** | Preserva bifurcações | Linear e limpo |
| **Segurança** | Seguro em branches públicas | ⚠️ Não usar em branches compartilhadas |
| **Rastreabilidade** | Mostra quando branches se uniram | Mais difícil de rastrear origem |

> **Regra:** Nunca faça rebase de branches que outras pessoas estão usando!

### 5.4 Resolvendo Conflitos

Conflitos acontecem quando duas branches modificam o **mesmo trecho** do mesmo arquivo.

**Arquivo com conflito:**
```python
def calcular_desconto(valor):
<<<<<<< HEAD (sua branch - main)
    return valor * 0.10
=======
    return valor * 0.15
>>>>>>> feature/desconto-vip
```

**Passo a passo para resolver:**

```bash
# 1. Git avisa sobre o conflito
# CONFLICT (content): Merge conflict in calculos.py

# 2. Abra o arquivo, edite manualmente e escolha o que manter:
def calcular_desconto(valor):
    return valor * 0.15  # ← versão escolhida (ou combinar as duas)

# 3. Marque o conflito como resolvido
git add calculos.py

# 4. Finalize o merge
git commit -m "merge: resolve conflito no cálculo de desconto"
```

> **Dica:** Use `git status` a qualquer momento para ver quais arquivos ainda têm conflitos.

## 6. GitHub — Repositório Remoto <a id="6"></a>

### O que é GitHub?

**GitHub** é uma plataforma de hospedagem de código baseada em Git. Além de armazenar repositórios, oferece:

- **Pull Requests** — proposta de merge com revisão de código
- **Issues** — rastreamento de bugs e tarefas
- **Actions** — CI/CD automatizado
- **Projects** — gerenciamento de projetos (Kanban)
- **Wiki** — documentação
- **Pages** — hospedagem de sites estáticos gratuita

### 6.1 Conectando Local ao Remoto

```bash
# Ver repositórios remotos configurados
git remote -v

# Adicionar repositório remoto
git remote add origin https://github.com/usuario/repo.git

# Renomear remote
git remote rename origin upstream

# Remover remote
git remote remove origin
```

### 6.2 Sincronizando com o Remoto

```
LOCAL                                    REMOTO (GitHub)
─────────────────────────────────────────────────────────

                    git push
  REPOSITORY  ─────────────────────►  REMOTE REPO

                    git fetch
  REPOSITORY  ◄─────────────────────  REMOTE REPO
  (atualiza refs, não muda working dir)

                    git pull
  WORKING DIR ◄─────────────────────  REMOTE REPO
  (fetch + merge automaticamente)
```

```bash
# Enviar commits para o remoto
git push origin main

# Primeiro push: vincular branch local à remota
git push -u origin main
# Após isso, basta: git push

# Baixar e integrar alterações do remoto
git pull origin main

# Baixar sem integrar (mais seguro)
git fetch origin
git diff main origin/main  # ver diferenças antes do merge
git merge origin/main      # integrar quando quiser

# Enviar nova branch para o remoto
git push -u origin feature/login

# Deletar branch no remoto
git push origin --delete feature/login
```

### 6.3 Fork e Contribuição Open Source

```
FLUXO DE CONTRIBUIÇÃO OPEN SOURCE:
────────────────────────────────────────────────────────────────

1. FORK: Copia o repo original para sua conta GitHub
   [repositorio/original]  →  [seu-usuario/original]

2. CLONE: Baixa seu fork para a máquina local
   git clone https://github.com/seu-usuario/original.git

3. UPSTREAM: Adiciona o repo original como referência
   git remote add upstream https://github.com/dono/original.git

4. BRANCH: Cria branch para sua contribuição
   git checkout -b fix/corrige-bug-login

5. CÓDIGO: Faz alterações, commits, push para seu fork
   git push origin fix/corrige-bug-login

6. PULL REQUEST: Abre PR do seu fork para o repo original

7. SINCRONIZAÇÃO: Mantém fork atualizado com o original
   git fetch upstream
   git merge upstream/main
────────────────────────────────────────────────────────────────
```

## 7. Gitflow — Fluxo de Trabalho Profissional <a id="7"></a>

### O que é Gitflow?

**Gitflow** é um modelo de ramificação (branching model) criado por **Vincent Driessen** em 2010. Define uma estrutura clara de branches para organizar o desenvolvimento de software.

É ideal para projetos com **ciclos de release definidos** (versões, deploys agendados).

### 7.1 As 5 Branches do Gitflow

```
TIPOS DE BRANCH NO GITFLOW:
────────────────────────────────────────────────────────────────────

  PERMANENTES (vivem para sempre):
  ┌─────────┐  Código em produção. Cada commit é uma versão lançada.
  │  main   │  Tag de versão a cada merge.
  └─────────┘

  ┌─────────┐  Integração de features. Código para a próxima versão.
  │ develop │  Base para criar features.
  └─────────┘

  TEMPORÁRIAS (criadas e deletadas):
  ┌───────────────────┐  Nova funcionalidade. Criada a partir de develop.
  │ feature/nome      │  Mesclada de volta em develop.
  └───────────────────┘

  ┌───────────────────┐  Preparação para release. Criada a partir de develop.
  │ release/v1.2.0    │  Permite ajustes finais sem travar develop.
  └───────────────────┘  Mesclada em main E develop.

  ┌───────────────────┐  Correção urgente em produção. Criada a partir de main.
  │ hotfix/bug-login  │  Mesclada em main E develop.
  └───────────────────┘
────────────────────────────────────────────────────────────────────
```

### 7.2 Diagrama Completo do Gitflow

```
TEMPO →

main     ●────────────────────────────────────●─────────────●
         v1.0                                 v1.1 (hotfix) v2.0
          \                                   /         \   /
           \                               (merge)    (merge)
develop     ●──●──●────────────●──●──●──●──●─────────────●
               /    \          /            \             /
              /   feature/B  (merge)      release/v2.0  /
feature/A   ●──●─/             \            ●──●──●─────
                               /
feature/B        ●──●──●──●──/

hotfix/                                     ●──●
bug-login                                  /     \
                                        (main)  (develop)

─────────────────────────────────────────────────────────────
LEGENDA:  ● = commit   / \ = branch/merge
```

### 7.3 Iniciando com Gitflow

```bash
# OPÇÃO 1: Manualmente (sem extensão)
git init
git commit --allow-empty -m "chore: initial commit"
git branch develop

# OPÇÃO 2: Com extensão git-flow (recomendado)

# Responder às perguntas (aceitar padrões com Enter)
```

### 7.4 Fluxo Completo — Exemplo Prático

**Cenário:** Time desenvolvendo um sistema de e-commerce.

---

#### PASSO 1: Feature — Desenvolver nova funcionalidade

```bash
# João vai implementar o carrinho de compras
git checkout develop
git pull origin develop               # sempre atualizar antes!

# Criar feature branch
git checkout -b feature/carrinho-compras
# ou com extensão:
git flow feature start carrinho-compras

# João desenvolve o código...
# (edita arquivos, testa, faz commits)
git add carrinho.py
git commit -m "feat: implementa adição de itens ao carrinho"
git add tests/test_carrinho.py
git commit -m "test: adiciona testes unitários do carrinho"

# Finalizar feature: merge em develop
git checkout develop
git merge --no-ff feature/carrinho-compras
# --no-ff = cria commit de merge mesmo se fast-forward for possível
#           preserva o histórico de que foi uma feature branch
git branch -d feature/carrinho-compras
git push origin develop
# ou com extensão:
git flow feature finish carrinho-compras
```

---

#### PASSO 2: Release — Preparar versão para produção

```bash
# Quando develop tem features suficientes para uma release
git checkout develop
git checkout -b release/v1.2.0
# ou:
git flow release start v1.2.0

# Na release branch: APENAS bugfixes, versioning, documentação
# (novas features vão para develop, não aqui!)
echo "1.2.0" > VERSION
git commit -am "chore: bump version to 1.2.0"

# Correção de bug encontrada durante testes
git commit -am "fix: corrige validação de CEP no checkout"

# Finalizar release: merge em main E develop
git checkout main
git merge --no-ff release/v1.2.0
git tag -a v1.2.0 -m "Release version 1.2.0"   # TAG na main!

git checkout develop
git merge --no-ff release/v1.2.0

git branch -d release/v1.2.0
git push origin main develop --tags
# ou:
git flow release finish v1.2.0
```

---

#### PASSO 3: Hotfix — Corrigir bug em produção

```bash
# BUG CRÍTICO em produção! Usuários não conseguem finalizar compra.

# Hotfix SEMPRE parte da main (código em produção)
git checkout main
git checkout -b hotfix/fix-finalizar-pedido
# ou:
git flow hotfix start fix-finalizar-pedido

# Corrigir o bug
git commit -am "fix: corrige erro ao finalizar pedido com desconto"

# Bump de versão patch
echo "1.2.1" > VERSION
git commit -am "chore: bump version to 1.2.1"

# Merge em main E develop (para não perder o fix!)
git checkout main
git merge --no-ff hotfix/fix-finalizar-pedido
git tag -a v1.2.1 -m "Hotfix 1.2.1"

git checkout develop
git merge --no-ff hotfix/fix-finalizar-pedido

git branch -d hotfix/fix-finalizar-pedido
git push origin main develop --tags
# ou:
git flow hotfix finish fix-finalizar-pedido
```

### 7.5 Resumo: Quando criar cada branch?

| Branch | Criada a partir de | Merge em | Quando usar |
|---|---|---|---|
| `feature/*` | `develop` | `develop` | Nova funcionalidade |
| `release/*` | `develop` | `main` + `develop` | Preparar próxima versão |
| `hotfix/*` | `main` | `main` + `develop` | Bug crítico em produção |
| `bugfix/*` | `develop` | `develop` | Bug encontrado durante dev |

### 7.6 Nomenclatura de Branches

```bash
# Padrão recomendado:
feature/nome-descritivo        # nova funcionalidade
feature/US-123-login-social    # com número da user story

release/v1.2.0                 # sempre com versão semântica

hotfix/descricao-curta         # descrição do bug
hotfix/BUG-456-erro-pagamento  # com número do ticket

bugfix/corrige-calculo-frete
chore/atualiza-dependencias
docs/adiciona-guia-instalacao
```

## 8. Pull Requests e Code Review <a id="8"></a>

### O que é um Pull Request (PR)?

Um **Pull Request** é uma proposta para mesclar código de uma branch em outra, com discussão e revisão antes do merge.

```
FLUXO DE UM PULL REQUEST:
─────────────────────────────────────────────────────────────────

1. Desenvolvedor faz push da feature branch
   git push origin feature/login

2. Abre Pull Request no GitHub:
   feature/login  →  develop

3. Revisores analisam o código:
   ✓ Lógica correta?
   ✓ Testes escritos?
   ✓ Padrões do projeto seguidos?
   ✓ Segurança? Performance?

4. Discussão e ajustes:
   Revisor comenta → Dev faz novos commits → Revisor aprova

5. Merge aprovado → branch deletada
─────────────────────────────────────────────────────────────────
```

### 8.1 Boas Práticas para PRs

**Como autor:**
- PRs pequenos e focados — mais fácil de revisar
- Título claro e descritivo
- Descrição explicando O QUÊ e POR QUÊ (não como)
- Screenshots/GIFs para mudanças visuais
- Referenciar issues relacionadas: `Closes #123`
- Todos os testes passando antes de abrir

**Como revisor:**
- Comentários construtivos e respeitosos
- Distinguir: sugestão opcional vs. bloqueante obrigatório
- Aprovar quando: código funciona, está correto, atende requisitos

### 8.2 Template de Pull Request

```markdown
## Descrição
Implementa autenticação via Google OAuth para facilitar o login dos usuários.

## Tipo de Mudança
- [x] Nova funcionalidade

- [ ] Documentação
```

### 9.1 Conventional Commits — Mensagens Padronizadas

**Conventional Commits** é uma especificação que define um formato padrão para mensagens de commit, tornando o histórico legível por humanos **e máquinas**.

> Para o guia completo e detalhado, veja a **[Seção 11 — Conventional Commits](#11)**.

**Formato básico:**
```
<tipo>(<escopo>): <descrição curta>
```

**Tipos mais usados:**

| Tipo | Dispara SemVer | Uso |
|---|---|---|
| `feat` | MINOR | Nova funcionalidade |
| `fix` | PATCH | Correção de bug |
| `feat!` / `fix!` | MAJOR | Breaking change |
| `docs` | — | Documentação |
| `refactor` | — | Refatoração sem mudança de comportamento |
| `test` | — | Testes |
| `chore` | — | Tarefas de manutenção |

**Exemplos rápidos:**

```bash
git commit -m "feat: adiciona filtro por categoria de produtos"
git commit -m "fix(cart): corrige contador de itens ao remover produto"
git commit -m "feat!: altera formato da resposta da API"
```

> **Benefício principal:** Com Conventional Commits é possível gerar **CHANGELOG automático** e calcular a próxima **versão semântica** de forma automatizada!

### 9.2 Versionamento Semântico (SemVer)

```
    v  1  .  2  .  3
       │     │     │
       │     │     └── PATCH: bug fix, hotfix (sem mudança de API)
       │     └──────── MINOR: nova funcionalidade (retrocompatível)
       └────────────── MAJOR: breaking change (incompatível)

Exemplos:
  v1.0.0 → v1.0.1   (hotfix)
  v1.0.1 → v1.1.0   (nova feature)
  v1.1.0 → v2.0.0   (breaking change)
```

### 9.3 Git Stash — Guardar Trabalho Temporariamente

```bash
# Situação: Estava trabalhando na feature, chegou urgência!
# Não quero commitar trabalho incompleto.

git stash                          # guarda mudanças
git stash push -m "WIP: tela de login"  # com descrição

# Resolve a urgência em outra branch...
git checkout hotfix/urgente
# ... faz as correções ...
git checkout feature/login

git stash list                     # ver o que está guardado
git stash pop                      # recupera e remove do stash
git stash apply stash@{0}          # recupera mas mantém no stash
git stash drop stash@{0}           # remove sem aplicar
git stash clear                    # limpa todo o stash
```

### 9.4 Git Tags — Marcando Versões

```bash
# Tag leve (só um ponteiro)
git tag v1.0.0

# Tag anotada (com metadados — RECOMENDADA para releases)
git tag -a v1.0.0 -m "Release versão 1.0.0 — lançamento inicial"

# Listar tags
git tag
git tag -l "v1.*"

# Enviar tags para o remoto
git push origin v1.0.0
git push origin --tags     # envia todas as tags

# Ver detalhes de uma tag
git show v1.0.0

# Deletar tag
git tag -d v1.0.0
git push origin --delete v1.0.0
```

### 9.6 GitHub Actions — CI/CD Automatizado

Exemplo de workflow que roda testes automaticamente a cada push:

```yaml
# .github/workflows/tests.yml
name: Testes Automáticos

on:
  push:
    branches: [ main, develop ]
  pull_request:
    branches: [ main, develop ]

jobs:
  test:
    runs-on: ubuntu-latest

    steps:
    - uses: actions/checkout@v3

    - name: Configurar Python
      uses: actions/setup-python@v4
      with:
        python-version: '3.11'

    - name: Instalar dependências
      run: |
        pip install -r requirements.txt

    - name: Rodar testes
      run: pytest tests/ -v

    - name: Verificar cobertura
      run: pytest --cov=src tests/ --cov-fail-under=80
```

> Com esse workflow, nenhum código com testes falhando entra na `main` ou `develop`!

## 11. Conventional Commits — Guia Completo <a id="11"></a>

### O que é e por que surgiu?

**Conventional Commits** é uma especificação criada para padronizar mensagens de commit. Surgiu da **Angular Commit Message Convention** (criada pelo time do Google para o framework Angular) e foi generalizada em 2019 como um padrão aberto.

**Problema que resolve:**

```
HISTÓRICO SEM PADRÃO (difícil de entender):
───────────────────────────────────────────
3f2a1c8  ajustes
9b4e5d2  correção
7a8c3f1  WIP
2d6e4b0  mais coisas
1a2b3c4  agora vai
0f9e8d7  tentativa 3
abc1234  FINALMENTE FUNCIONOU

HISTÓRICO COM CONVENTIONAL COMMITS (informativo):
───────────────────────────────────────────────────
3f2a1c8  feat(auth): implementa login com OAuth do Google
9b4e5d2  fix(cart): corrige contador ao remover produto duplicado
7a8c3f1  refactor(api): extrai lógica de validação para serviço
2d6e4b0  test(auth): adiciona testes de integração para OAuth
1a2b3c4  docs: atualiza README com instrução de configuração OAuth
0f9e8d7  chore(deps): atualiza requests para 2.31.0
abc1234  feat!: altera retorno da API para formato JSON:API
```

O segundo histórico permite:
- Entender cada mudança **sem abrir o código**
- Gerar **CHANGELOG automático**
- Determinar **próxima versão** automaticamente
- Filtrar commits por tipo (`git log --grep="^feat"`)
- **Automatizar releases** completas

### 11.1 Anatomia Completa de um Commit

Um commit Conventional tem até **três partes**:

```
┌─────────────────────────────────────────────────────────────────────┐
│  LINHA 1 — HEADER (obrigatório)                                     │
│                                                                     │
│  feat(auth): implementa autenticação com Google OAuth               │
│  ──┬─  ──┬─  ──────────────────────────────────────                │
│    │     │                      │                                   │
│    │     │                      └── descrição: curta, imperativa,   │
│    │     │                          minúscula, sem ponto no final   │
│    │     └── escopo (opcional): qual parte do sistema foi afetada   │
│    └──────── tipo: categoria da mudança                             │
│                                                                     │
│  (linha em branco obrigatória entre header e body)                  │
│                                                                     │
│  CORPO — BODY (opcional)                                            │
│                                                                     │
│  Adiciona autenticação via OAuth 2.0 usando a biblioteca            │
│  google-auth. O usuário pode optar por login tradicional ou         │
│  Google. Tokens são armazenados de forma segura via httpOnly        │
│  cookies com expiração de 24 horas.                                 │
│                                                                     │
│  (linha em branco obrigatória entre body e footer)                  │
│                                                                     │
│  RODAPÉ — FOOTER (opcional, pode ter múltiplos)                     │
│                                                                     │
│  BREAKING CHANGE: o endpoint /login agora requer campo 'provider'  │
│  Closes #102                                                        │
│  Co-authored-by: Maria Silva <maria@email.com>                      │
└─────────────────────────────────────────────────────────────────────┘
```

**Regras do Header:**

| Parte | Obrigatório | Regras |
|---|---|---|
| `tipo` | Sim | Letras minúsculas; da lista definida pelo projeto |
| `(escopo)` | Não | Entre parênteses; minúsculo; sem espaços |
| `!` | Não | Indica breaking change; vai após escopo ou tipo |
| `:` + espaço | Sim | Separador obrigatório entre tipo e descrição |
| `descrição` | Sim | Imperativo, minúscula, sem ponto final, máx ~72 chars |

### 11.2 Tipos — Referência Completa

| Tipo | Impacto SemVer | Aparece no CHANGELOG | Descrição |
|---|---|---|---|
| `feat` | **MINOR** (1.x.0) | Sim | Nova funcionalidade para o usuário final |
| `fix` | **PATCH** (1.0.x) | Sim | Correção de bug que afeta o usuário |
| `perf` | **PATCH** | Sim | Melhoria de performance perceptível |
| `revert` | PATCH | Sim | Reverte um commit anterior |
| `refactor` | — | Não | Reestruturação interna sem mudança de comportamento |
| `style` | — | Não | Formatação, espaços, vírgulas (sem lógica alterada) |
| `test` | — | Não | Adiciona ou corrige testes |
| `docs` | — | Não | Mudanças apenas na documentação |
| `build` | — | Não | Sistema de build, dependências externas (npm, pip) |
| `ci` | — | Não | Configuração de CI/CD (Actions, Jenkins, CircleCI) |
| `chore` | — | Não | Outras tarefas que não afetam src nem tests |
| **qualquer`!`** | **MAJOR** (x.0.0) | Sim | Breaking change (qualquer tipo pode ter `!`) |

> **Dica:** Os únicos tipos que **obrigatoriamente** entram no CHANGELOG são `feat`, `fix`, `perf` e qualquer commit com breaking change. Os demais são opcionais por ferramenta.

**Exemplos de cada tipo:**

```bash
feat(pedidos):  adiciona opção de parcelamento em até 12x
fix(auth):      corrige token não renovado após expiração
perf(db):       adiciona índice na coluna user_id da tabela pedidos
revert:         reverte "feat: adiciona tema escuro"
refactor(api):  extrai classe ApiClient para módulo separado
style:          aplica formatação black em todos os arquivos
test(pagamento): adiciona testes para fluxo de estorno
docs(api):      adiciona exemplos de uso no README da API
build(deps):    atualiza Django de 4.1 para 4.2
ci:             adiciona cache de dependências no workflow
chore:          remove arquivos temporários do diretório /tmp
```

### 11.3 Escopo (Scope)

O escopo indica **qual parte do sistema** foi modificada. É opcional, mas muito recomendado em projetos maiores.

**Convenções para escopos:**

```bash
# Por módulo/domínio do sistema
feat(auth):      # módulo de autenticação
feat(pedidos):   # módulo de pedidos
fix(pagamento):  # módulo de pagamento
fix(relatorio):  # módulo de relatórios

# Por camada da arquitetura
refactor(api):   # camada de API/controllers
test(service):   # camada de serviços
fix(model):      # camada de modelos/domínio
chore(db):       # banco de dados/migrations

# Por tipo de componente
feat(ui):        # interface do usuário
fix(cli):        # interface de linha de comando
docs(readme):    # arquivo README específico

# Por arquivo/componente específico
fix(carrinho):   # componente de carrinho
test(usuario):   # testes relacionados a usuário
```

**Boas práticas para escopos:**
- Use **substantivos** (não verbos): `auth`, não `autenticar`
- Mantenha um **glossário de escopos** no projeto (CONTRIBUTING.md)
- Seja **consistente** — o mesmo módulo sempre usa o mesmo escopo
- Use **minúsculas** e **sem espaços** (use hífen se necessário: `lista-espera`)

### 11.4 Descrição (Subject)

A descrição é o **resumo da mudança** na primeira linha. É a parte mais lida do commit.

**Regras obrigatórias:**
```
✓ Modo imperativo: "adiciona", "corrige", "remove" (não "adicionado", "corrigindo")
✓ Minúscula na primeira letra
✓ Sem ponto final
✓ Máximo ~72 caracteres (para não truncar no terminal/GitHub)
✓ Em português OU inglês — escolha um e seja consistente
```

**Modo imperativo — como pensar:**

> "Se aplicado, este commit vai... **[descrição]**"

```bash
# CORRETO — modo imperativo
feat: adiciona validação de CPF no cadastro
fix:  remove loop infinito no processamento de pagamento
docs: inclui exemplos de uso na documentação da API

# ERRADO
feat: adicionando validação de CPF     ← gerúndio
feat: Validação de CPF adicionada      ← passado + maiúscula
feat: valida o CPF.                    ← ponto no final
feat: adiciona validação de CPF no campo de cadastro de novos usuários
     ← muito longo (>72 chars)
```

**Dica para escrever bem a descrição:**

```
Ruim:  "fix: correção"           → O quê foi corrigido? Onde? Por quê?
Ruim:  "feat: changes"           → Mudanças em quê?
Ruim:  "chore: misc updates"     → Inútil
Bom:   "fix(auth): corrige sessão expirada não redirecionando para login"
Bom:   "feat(relatorio): adiciona exportação de dados para CSV"
```

### 11.5 Corpo (Body)

O corpo explica o **contexto e motivação** da mudança. Use quando o header não é suficiente.

**Quando usar o corpo:**
- A mudança é complexa e merece explicação
- Você precisa detalhar **o que mudou** e **por que** (não como)
- Há trade-offs ou decisões arquiteturais para registrar

```bash
# Commit com corpo completo
git commit -m "fix(pagamento): corrige cálculo de juros no parcelamento

O cálculo anterior usava juros simples para todas as parcelas,
resultando em valores incorretos para parcelamentos acima de 6x.
A correção aplica juros compostos usando a fórmula de Price.

Afeta: todos os pedidos com parcelamento em 7 ou mais vezes.
Pedidos anteriores NÃO são recalculados retroativamente."
```

**Regras do corpo:**
- Separado do header por **uma linha em branco**
- Sem limite rígido de linhas (mas 72 chars por linha é convencional)
- Explique o **POR QUÊ**, não o como (o código explica o como)
- Pode usar listas com `-` ou `*`

```bash
# Corpo com lista
git commit -m "refactor(auth): separa responsabilidades do AuthService

Motivação: o AuthService estava com mais de 400 linhas e múltiplas
responsabilidades (autenticação, autorização, sessões, tokens).

Mudanças realizadas:
- Extrai TokenService para gerenciar JWT e refresh tokens
- Extrai SessionService para gerenciar sessões de usuário
- AuthService agora apenas coordena os dois serviços acima
- Mantém 100% de compatibilidade com a interface pública existente"
```

### 11.6 Rodapé (Footer)

O rodapé contém **metadados estruturados** do commit. Cada entrada segue o formato `Token: valor`.

**Tokens padrão:**

| Token | Uso | Exemplo |
|---|---|---|
| `BREAKING CHANGE` | Descreve uma mudança incompatível | `BREAKING CHANGE: remove suporte a Python 3.8` |
| `Closes` / `Fixes` | Fecha uma issue automaticamente | `Closes #123` |
| `Refs` | Referencia sem fechar | `Refs #456` |
| `Co-authored-by` | Credita co-autor (pair programming) | `Co-authored-by: Ana <ana@email.com>` |
| `Reviewed-by` | Quem revisou o código | `Reviewed-by: Pedro <pedro@email.com>` |

```bash
# Commit com múltiplos footers
git commit -m "feat(api): adiciona endpoint de exportação em lote

Permite exportar até 10.000 registros em formato CSV ou XLSX.
Processamento é feito de forma assíncrona via Celery para não
bloquear a requisição HTTP.

BREAKING CHANGE: o endpoint GET /export foi renomeado para
GET /export/single. O novo endpoint de lote é GET /export/batch.

Closes #234
Closes #198
Co-authored-by: Carlos Lima <carlos@empresa.com>
Reviewed-by: Ana Souza <ana@empresa.com>"
```

**Como o GitHub interpreta os tokens:**
- `Closes #N` fecha a issue N automaticamente quando o PR é mergeado
- `Co-authored-by` aparece no perfil do co-autor como contribuição
- `BREAKING CHANGE` é detectado por ferramentas como `semantic-release`

### 11.7 Breaking Changes — Em Detalhes

Um **breaking change** é qualquer mudança que **quebra compatibilidade** com versões anteriores — usuários da API/biblioteca precisarão ajustar seu código.

**Duas formas de declarar:**

```bash
# FORMA 1: ! no header (atalho, apenas sinaliza)
feat!: remove suporte ao formato XML na API
fix(auth)!: altera algoritmo de hash de MD5 para bcrypt

# FORMA 2: BREAKING CHANGE no footer (com descrição completa — RECOMENDADO)
git commit -m "feat(api): migra autenticação de API Key para OAuth 2.0

Implementa fluxo OAuth 2.0 com suporte a client_credentials
e authorization_code. Melhora segurança e rastreabilidade.

BREAKING CHANGE: autenticação via header 'X-API-Key' foi removida.
Clientes devem migrar para Bearer Token OAuth 2.0.
Guia de migração: https://docs.empresa.com/migration/oauth"

# FORMA 3: Combinar ! e BREAKING CHANGE (mais explícito)
git commit -m "feat(api)!: migra autenticação para OAuth 2.0

BREAKING CHANGE: autenticação via header 'X-API-Key' foi removida."
```

**O que caracteriza um breaking change?**

```
BREAKING CHANGE — EXEMPLOS:
───────────────────────────────────────────────────────────────────
✗ Remover um campo de resposta da API
✗ Renomear um endpoint (ex: /users → /usuarios)
✗ Alterar o tipo de um campo (string → integer)
✗ Tornar obrigatório um campo que era opcional
✗ Alterar o comportamento padrão de uma função
✗ Remover suporte a versão de linguagem/framework
✗ Alterar formato de retorno (JSON → XML, ou estrutura interna)

NÃO é breaking change:
✓ Adicionar novo campo OPCIONAL na resposta
✓ Adicionar novo endpoint
✓ Adicionar parâmetro opcional com valor padrão
✓ Melhorar performance sem alterar interface
───────────────────────────────────────────────────────────────────
```

### 11.8 Relação com Versionamento Semântico (SemVer)

Conventional Commits e SemVer foram feitos para trabalhar juntos. Cada tipo de commit **determina automaticamente** qual dígito da versão incrementar:

```
COMMIT                              VERSÃO
──────────────────────────────────────────────────────────────────
fix: corrige bug no login           v1.2.3  →  v1.2.4  (PATCH ++)
feat: adiciona exportação CSV       v1.2.4  →  v1.3.0  (MINOR ++)
feat!: remove endpoint legado       v1.3.0  →  v2.0.0  (MAJOR ++)

Múltiplos commits no mesmo release:
fix + fix + feat                    →  MINOR (feat prevalece sobre fix)
fix + fix + feat!                   →  MAJOR (breaking change prevalece)
──────────────────────────────────────────────────────────────────
```

**Regra de precedência (maior impacto vence):**
```
MAJOR > MINOR > PATCH
  !   >  feat  >  fix/perf/revert
```

**Visualização completa:**

```
                    BREAKING CHANGE ou !
                            │
                            ▼
feat ──────────────────────────────────────► MINOR bump
fix, perf, revert ─────────────────────────► PATCH bump
refactor, style, test, docs, build, ci ────► sem bump
chore ──────────────────────────────────────► sem bump

Nota: qualquer tipo com ! ou BREAKING CHANGE → MAJOR bump
      independente do tipo base
```

```
╚════════════════════════════════════╩═══════════════════════════════════════════╝
```

**Princípio do commit atômico:**

> Um commit deve representar **uma única mudança lógica**. Se você usa "e" na descrição, provavelmente são dois commits.

```bash
# Errado: commit com responsabilidades misturadas
git commit -m "feat: adiciona carrinho e corrige bug no login e atualiza README"

# Correto: commits separados, cada um com propósito único
git commit -m "feat(carrinho): implementa adição e remoção de itens"
git commit -m "fix(auth): corrige redirecionamento após login expirado"
git commit -m "docs: adiciona seção de instalação no README"
```

### 11.11 Ferramentas do Ecossistema

#### Commitizen — Interface interativa para criar commits

Guia o desenvolvedor pelo formato correto com perguntas:

```bash
# Instalar (Python)
pip install commitizen

# Inicializar no projeto
cz init

# Criar commit interativamente (substitui git commit)
cz commit
# ou
git cz
```

**Exemplo de interação com Commitizen:**
```
? Select the type of change you are committing:
  feat:     A new feature
  fix:      A bug fix
❯ refactor: A code change that neither fixes a bug nor adds a feature
  ...

? What is the scope of this change? (press enter to skip)
> auth

? Write a short, imperative tense description of the change:
> extrai lógica de validação para AuthValidator

? Provide a longer description of the change: (press enter to skip)
> Separa responsabilidade de validação do AuthService

? Are there any breaking changes? No
? Does this change affect any open issues? No

→ Commit: refactor(auth): extrai lógica de validação para AuthValidator
```

---

#### Commitlint — Validação automática de commits

Rejeita commits que não seguem o padrão:

```bash
# Instalar (Node.js/npm no projeto)
npm install --save-dev @commitlint/cli @commitlint/config-conventional

# Configurar (commitlint.config.js)
module.exports = { extends: ['@commitlint/config-conventional'] }
```

```
# Exemplos de rejeição:
$ git commit -m "ajustes"
⧗  input: ajustes
✖  subject may not be empty [subject-empty]
✖  type may not be empty [type-empty]
✖  found 2 problems, 0 warnings

$ git commit -m "feat: Adiciona Login."
⧗  input: feat: Adiciona Login.
✖  subject must not be sentence-case [subject-case]
✖  subject may not end with full stop [subject-full-stop]
```

---

#### Husky — Git Hooks automatizados

Executa validações **antes** de cada commit:

```bash
# Instalar
npm install --save-dev husky
npx husky init

# .husky/commit-msg (hook que valida a mensagem)
npx --no -- commitlint --edit $1
```

Com Husky + Commitlint, mensagens inválidas são **bloqueadas automaticamente** antes de entrar no repositório.

---

#### semantic-release — Release totalmente automatizada

Analisa os commits, determina a versão, publica no npm/PyPI e cria o CHANGELOG:

```bash
# Instalar
npm install --save-dev semantic-release

# Com semantic-release configurado:
# 1. Developer faz push para main com commits convencionais
# 2. CI/CD executa semantic-release automaticamente
# 3. semantic-release:
#    - Analisa commits desde última tag
#    - Determina nova versão (MAJOR/MINOR/PATCH)
#    - Cria tag Git (v1.3.0)
#    - Publica pacote no npm/PyPI
#    - Gera CHANGELOG
#    - Cria GitHub Release com notas automáticas
```

---

#### git-cliff — Gerador de CHANGELOG em Rust (muito rápido)

```bash
# Instalar
cargo install git-cliff
# ou: pip install git-cliff (wrapper Python)

# Gerar CHANGELOG desde o início do projeto
git cliff --output CHANGELOG.md

# Gerar apenas para a próxima versão
git cliff --unreleased --tag v2.0.0
```

```
║                                                                      ║

║  test      → (nada)  Adição/correção de testes                      ║

║  build     → (nada)  Sistema de build / dependências                ║

╠══════════════════════════════════════════════════════════════════════╣

╠══════════════════════════════════════════════════════════════════════╣

```

> **Conventional Commits** transforma mensagens de commit de texto livre em **dados estruturados**, habilitando automação completa do ciclo de release: versão, changelog e publicação, tudo sem intervenção manual.
