## 5. Branches вҖ” RamificaГ§Гөes do Projeto <a id="5"></a>

### O que Г© uma Branch?

Uma **branch** (ramificaГ§ГЈo) Г© uma linha de desenvolvimento independente. Ela permite trabalhar em funcionalidades, correГ§Гөes ou experimentos **sem afetar o cГіdigo principal**.

```

                                (merge na main)
```

Cada letra representa um commit. A branch `feature` foi criada a partir de `B`, desenvolvida independentemente, e depois mesclada de volta Г  `main` no ponto `M`.

### 5.1 Comandos de Branch

```bash
# Listar branches locais

# Listar todas (locais + remotas)

# Criar nova branch

# Trocar para a branch

# Criar e jГЎ trocar (atalho)

# ou

# Deletar branch (apГіs merge)

# ForГ§ar deleГ§ГЈo (sem merge)

```

### 5.2 Merge вҖ” Unindo Branches


### 5.4 Resolvendo Conflitos

Conflitos acontecem quando duas branches modificam o **mesmo trecho** do mesmo arquivo.

**Arquivo com conflito:**
```python

<<<<<<< HEAD (sua branch - main)
    return valor * 0.10
=======
    return valor * 0.15
>>>>>>> feature/desconto-vip
```

**Passo a passo para resolver:**

```bash

# 2. Abra o arquivo, edite manualmente e escolha o que manter:

    return valor * 0.15  # вҶҗ versГЈo escolhida (ou combinar as duas)

# 3. Marque o conflito como resolvido

# 4. Finalize o merge

```

> **Dica:** Use `git status` a qualquer momento para ver quais arquivos ainda tГӘm conflitos.
